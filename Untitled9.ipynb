```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("/content/Salary.xlsx")
data.head()
```

```python
df=pd.DataFrame(data) #Crear un dataframe
df.head()
```

```python
df.columns
```

```python
X=df["YearsExperience"]  #variable independiente
Y=df['Salary']  #variable dependiente

plt.figure(figsize=(10,6))
plt.scatter(X,Y, color="blue", label="Datos reales")
plt.title("Salario vs AÃ±os de experiencia")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Ajustando los datos a un modelo de regresion lineal
model=LinearRegression()
model.fit(X.values.reshape(-1,1),Y)

#Generar predicciones
y_pred=model.predict(X.values.reshape(-1,1))

#Hallar el R2
r2=r2_score(Y,y_pred)
```

```python
plt. figure(figsize=[10, 6])
plt.scatter (X, Y, color="blue",label="Datos Reales")
plt.plot(X,y_pred, color="red", linewidth=3, label= "RegresiÃ³n Lineal")
plt.xlabel("AÃ±os de experiencia")
plt. ylabel("Salario")
plt. title("Salario vs AÃ±os de experiencia")
plt. legend()
plt.grid (True)
plt. show()

print(f"Ecuacion de la recta: y={model.coef_[0]:.2f}x+ {model.intercept_:2f}")
print(f"Coeficiente de determinacion (R^2):{r2:.2f}")
```
