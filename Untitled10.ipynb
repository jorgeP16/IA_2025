```python
def bar_chart(feature):
  survived=df[df['Survived']==1][feature].value_counts()
  dead=df[df['Survived']==0][feature].value_counts()
  test=pd.DataFrame([survived,dead])
  test.index=['Survived','Dead']
  test.plot(kind='bar',stacked=True,figsize=(10,5))
```

```python
median_bedrooms = df['total_bedrooms'].median()
df['total_bedrooms'].fillna(median_bedrooms, inplace=True)
missing_after_fill = df['total_bedrooms'].isnull().sum()
print(f"Number of missing values in 'total_bedrooms' after filling: {missing_after_fill}")
```

```python
df['bedrooms_per_room'] = df['total_bedrooms'] / df['total_rooms']
df['age'] = 2023 - df['housing_median_age']
df['rooms_per_household'] = df['total_rooms'] / df['households']
df['population_per_household'] = df['population'] / df['households']
display(df.head())
```

```python
four_bedroom_houses = df[df['total_bedrooms'] == 4]
display(four_bedroom_houses.head())
print(f"Number of four-bedroom houses: {len(four_bedroom_houses)}")
```

```python
numerical_features = ['age', 'median_income', 'total_rooms', 'population', 'bedrooms_per_room', 'rooms_per_household', 'population_per_household']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_features):
    plt.subplot(3, 3, i + 1)
    sns.scatterplot(x=feature, y='median_house_value', data=four_bedroom_houses)
    plt.title(f'Median House Value vs {feature}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='ocean_proximity', y='median_house_value', data=four_bedroom_houses)
plt.title('Median House Value vs Ocean Proximity')
plt.show()
```

```python
correlation_matrix = four_bedroom_houses[numerical_features + ['median_house_value']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de correlacin de caractersticas numricas para casas de cuatro dormitorios')
plt.show()
```

```python
ocean_proximity_ranking = four_bedroom_houses.groupby('ocean_proximity')['median_house_value'].median().sort_values(ascending=False)
ocean_proximity_map = {category: rank for rank, category in enumerate(ocean_proximity_ranking.index)}

four_bedroom_houses['ocean_proximity_score'] = four_bedroom_houses['ocean_proximity'].map(ocean_proximity_map)

for feature in ['median_income', 'rooms_per_household', 'bedrooms_per_room', 'age']:
    min_val = four_bedroom_houses[feature].min()
    max_val = four_bedroom_houses[feature].max()
    if feature == 'bedrooms_per_room':
        four_bedroom_houses[f'{feature}_scaled'] = (max_val - four_bedroom_houses[feature]) / (max_val - min_val)
    else:
        four_bedroom_houses[f'{feature}_scaled'] = (four_bedroom_houses[feature] - min_val) / (max_val - min_val)

four_bedroom_houses['quality_score'] = (
    four_bedroom_houses['median_income_scaled'] +
    four_bedroom_houses['rooms_per_household_scaled'] +
    four_bedroom_houses['ocean_proximity_score'] - four_bedroom_houses['bedrooms_per_room_scaled'] # Subtracting as lower is better
)

four_bedroom_houses['overall_score'] = four_bedroom_houses['quality_score'] + four_bedroom_houses['age_scaled']

ranked_houses = four_bedroom_houses.sort_values(by='overall_score', ascending=False)

display(ranked_houses[['median_house_value', 'median_income', 'total_rooms', 'total_bedrooms', 'age', 'ocean_proximity', 'quality_score', 'overall_score']])
```

```python
print("Las casas mejor clasificadas:")
display(ranked_houses.head())

best_house = ranked_houses.iloc[0]
print("\nCaractersticas de la mejor casa:")
display(best_house[['median_house_value', 'median_income', 'total_rooms', 'total_bedrooms', 'age', 'ocean_proximity', 'quality_score', 'overall_score']])
```
